```python
#Total sum of goals by each Club.
tg=epl.groupby('Club')['Goals'].sum()
tg.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

tg.plot(kind='barh',cmap='summer')
plt.xlabel('Goals')
plt.title('Goals By Clubs', fontsize=17)
plt.ylabel('Clubs');
```

```python
club_assit=epl.groupby('Club')['Assists'].sum()

club_assit
```

```python
fig, ax = plt.subplots(figsize=(10, 8))


club_assit.plot( kind='barh',color='indigo')

plt.title('Assists By Clubs', fontsize=18)
plt.ylabel('Assists');


```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(x='Club',y='Age',data=epl)
plt.xlabel("Club",fontsize=25)
plt.ylabel("Age",fontsize=20)
plt.xticks(rotation=90);
plt.title('Age Range of Players');
```

```python
Avg_age=epl.groupby('Club')['Age'].mean().plot(kind='bar')
```

```python

fig, ax = plt.subplots(figsize=(10, 8))


Avg_age.plot( kind='barh',color='green')

plt.title('Average Ages of Players by Club', fontsize=18)
plt.xlabel('Average Age')
plt.ylabel('Clubs');
```

```python
gg=epl[['Name','Goals','Assists']].nlargest(10,'Goals').reset_index(drop=True).sort_values(by='Assists')
gg
```

```python
plot=plt.barh(gg.Name,gg.Assists)
for rectangle in plot:
    width = rectangle.get_width()
    plt.text(width + 0.3, rectangle.get_y() + 0.16,'%d'%int(width),
    ha='center', va = 'bottom')
plt.xlabel('Assist')
plt.title('Assist Kings')
plt.show()

```
